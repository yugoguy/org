```python
import pandas as pd
import numpy as np
import pickle
import warnings
import requests
warnings.filterwarnings("ignore")
```

```python
latest_pagenum = 542
```

```python
update = range(2,latest_pagenum+1)
```

```python
htmlraw = requests.get(f"https://www.yfull.com/snp-list/?page={1}")
content = pd.read_html(htmlraw.content, header=0)
YFull = content[1]

YFull
```

```python
for page in update:
    htmlraw = requests.get(f"https://www.yfull.com/snp-list/?page={page}")
    content = pd.read_html(htmlraw.content, header=0)
    current = content[1]
    YFull = pd.concat([YFull,current])
    print(page)
```

```python
YFull
```

```python
YFull.to_pickle("YFull_updates.sav")
```
